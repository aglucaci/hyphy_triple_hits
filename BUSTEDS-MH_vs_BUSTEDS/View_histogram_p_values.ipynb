```python
import pandas as pd
import plotly.express as px
from prettytable import PrettyTable
```

```python
#https://pandas.pydata.org/pandas-docs/version/0.23.1/generated/pandas.read_csv.html#pandas.read_csv
df = pd.read_csv("processed_BUSTEDS-MH_vs_BUSTEDS.csv")
```

```python
len(df.index)
```

```python



#Stats
thresholds = [0.05, 0.01, 0.005, 0.001]

x = PrettyTable()

x.field_names = ["Model", 0.05, 0.01, 0.005, 0.001]


add_to_table = ["BUSTEDSMH"] 
for thresh in thresholds:
    #print(thresh)
    count = 0
    for item in df["pvalue_BUSTEDSMH"]:
        #print(item, type(item))
        if item <= thresh:
            count += 1
        #end if
    #end inner for
    print("For threshold:", thresh, "passed =", count/len(df.index) * 100)
    add_to_table.append(str(round(count/len(df.index) * 100,1)) + "%")
#end outer for 

x.add_row(add_to_table)



add_to_table = ["BUSTEDS"] 
for thresh in thresholds:
    #print(thresh)
    count = 0
    for item in df["pvalue_BUSTEDS"]:
        #print(item, type(item))
        if item <= thresh:
            count += 1
        #end if
    #end inner for
    print("For threshold:", thresh, "passed =", count/len(df.index) * 100)
    add_to_table.append(str(round(count/len(df.index) * 100,1)) + "%")
#end outer for 

x.add_row(add_to_table)
print()

print(x)
```

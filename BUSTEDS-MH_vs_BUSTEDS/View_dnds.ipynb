```python
import pandas as pd
import plotly.express as px
from prettytable import PrettyTable
import plotly.graph_objects as go
import numpy as np
```

```python
#https://pandas.pydata.org/pandas-docs/version/0.23.1/generated/pandas.read_csv.html#pandas.read_csv
df = pd.read_csv("processed_BUSTEDS-MH_vs_BUSTEDS_with_dnds.csv")
```

```python
len(df.index)
```

```python
#Stats
thresholds = [0.05, 0.01, 0.005, 0.001]
#Model 1
x = PrettyTable()
x.field_names = ["Model", 0.05, 0.01, 0.005, 0.001]
add_to_table = ["BUSTEDSMH"] 
for thresh in thresholds:
    #print(thresh)
    count = 0
    for item in df["pvalue_BUSTEDSMH"]:
        #print(item, type(item))
        if item <= thresh:
            count += 1
        #end if
    #end inner for
    print("\tCounts SMH:", count)
    print("For threshold:", thresh, "passed =", count/len(df.index) * 100)
    print()
    add_to_table.append(str(round(count/len(df.index) * 100,1)) + "%")
#end outer for 

x.add_row(add_to_table)

# Model 2
add_to_table = ["BUSTEDS"] 
for thresh in thresholds:
    #print(thresh)
    count = 0
    for item in df["pvalue_BUSTEDS"]:
        #print(item, type(item))
        if item <= thresh:
            count += 1
        #end if
    #end inner for
    print("\tCounts SRV:", count)
    print("For threshold:", thresh, "passed =", count/len(df.index) * 100)
    add_to_table.append(str(round(count/len(df.index) * 100,1)) + "%")
#end outer for 

x.add_row(add_to_table)
print()

print(x)
```

```python
import plotly.graph_objects as go

import numpy as np

x0 = np.random.randn(500)
# Add 1 to shift the mean of the Gaussian distribution
x1 = np.random.randn(500) + 1

fig = go.Figure()
fig.add_trace(go.Histogram(x=df["pvalue_BUSTEDS"], name='BUSTEDS'))
fig.add_trace(go.Histogram(x=df["pvalue_BUSTEDSMH"], name='BUSTEDS-MH'))

# Overlay both histograms
fig.update_layout(barmode='overlay')
# Reduce opacity to see both histograms
fig.update_traces(opacity=0.75)

fig.update_layout(
    title_text='p-value distribution for BUSTEDS and BUSTEDSMH', # title of plot
    xaxis_title_text='Value', # xaxis label
    yaxis_title_text='Count') # yaxis label)

fig.update_traces(marker_line_color='rgb(8,48,107)',
                  marker_line_width=1.5, opacity=0.75)

fig.show()
```
